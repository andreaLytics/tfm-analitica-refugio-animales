## Análisis exploratorio inicial del dataset
Objetivo: comprender la estructura, calidad y variables del dataset
procedente del Austin Animal Center.

```python
import pandas as pd
```

```python
datos = pd.read_csv("aac_intakes_outcomes.csv")
```

**Eliminar columnas**:

Las columnas "animal_id_outcome" y "animal_id_intake" tienen los mismos valores ya que el animal solo genera un id. Eliminaremos una.

Las columnas "outcome_number" y "intake_number" tienen los mismos valores ya que es el número de veces que animal ha estado en el refugio. Eliminaremos una.


```python
cols_to_drop = [
    "animal_id_outcome",
    "outcome_number"
]

datos2 = datos.drop(columns=cols_to_drop)
```

```python
datos2.head()
```

**Eliminar posibles duplicados**:

```python
datos2.drop_duplicates(inplace=True)
```

**Sustituir valores desconocidos:**

La única variable con un porcentaje de valores desconocidos es "outcome_subtype" con un 54%. Se sustituirán con la palabra "Unknown".


```python
datos2['outcome_subtype'] = datos2['outcome_subtype'].fillna('Unknown')
```

**Normalizamos texto**:

```python
cols_to_strip = ["animal_type", "breed", "sex_upon_outcome", "color", "found_location", "intake_condition", "intake_type", "sex_upon_intake", "outcome_type", "outcome_subtype", "outcome_weekday", "intake_weekday"]
for col in cols_to_strip:
    datos2[col] = datos2[col].str.strip().str.title()
```

```python
datos2.head()
```

**Convertir fechas a datetime y extraer solo la fecha**:

```python
datos2['intake_datetime'] = pd.to_datetime(datos2['intake_datetime'])
datos2['outcome_datetime'] = pd.to_datetime(datos2['outcome_datetime'])
datos2['intake_date'] = datos2['intake_datetime'].dt.date
datos2['outcome_date'] = datos2['outcome_datetime'].dt.date
```

```python
datos2.head()
```

**Creación de indicadores binarios**:

```python
datos2['outcome_type'].value_counts()
```

```python
datos2['IsAdopted'] = (datos2['outcome_type'] == 'Adoption').astype(int)
datos2['IsEuthanized'] = (datos2['outcome_type'] == 'Euthanasia').astype(int)
datos2['IsReturned'] = datos2['outcome_type'].str.contains(
    'return to owner|rto',
    case=False,
    na=False
).astype(int)
datos2['IsSick'] = (datos2['intake_condition'] == 'Sick').astype(int)
datos2['IsInjured'] = (datos2['intake_condition'] == 'Injured').astype(int)
datos2['IsRescue'] = (datos2['intake_type'] == 'Stray').astype(int)
datos2["Reentry"] = (datos2["intake_number"] > 1).astype(int)
datos2["Istransfer"] = (datos2["outcome_type"] == "Transfer").astype(int)
```
